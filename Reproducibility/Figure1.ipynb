```python
import scanpy as sc
import pandas as pd
import seaborn as sn
import statsmodels.api as sm
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
sc.settings.verbosity = 1
sc.set_figure_params(vector_friendly=True,dpi_save=300,transparent=True)
```

# Figure 1f

```python
lupus = sc.read("./data/lupus/lupus.txt",sep="\t",cache=True).T
lupus.obs = pd.read_csv("./data/lupus/lupus_obs.txt",sep="\t")
sc.pp.filter_genes(lupus,min_cells=3)
X = lupus.obs.loc[:,['tsne1','tsne2']]
X = StandardScaler().fit_transform(X)
db = DBSCAN(eps=.1, min_samples=20).fit(X)
lupus.obs['labels'] = db.labels_
lupus_cleaned = lupus[lupus.obs['labels']==0,]
sn.scatterplot(data=lupus_cleaned.obs,x='tsne1',y='tsne2',hue='Celltype')
```

# Figure 1j

```python
lupus_cleaned = lupus[lupus.obs['labels']==0,]
lupus_cleaned_sub = sc.pp.subsample(lupus_cleaned,fraction=0.1,copy=True)
sc.pp.filter_genes(lupus_cleaned_sub,min_cells=10)
sc.pp.log1p(lupus_cleaned_sub)
sc.pp.scale(lupus_cleaned_sub)
```

```python
def part1(gene_set,tf_name,group,result,count):
    targets = gene_set[tf_name]
    targets = list(set(targets) & set(count.var_names))
    c = count[:,targets].obs.index.tolist()
    coord = pd.DataFrame(result[0][tf_name])
    coord = coord.set_index(lupus_cleaned.obs.index)
    coord = coord.loc[c,:]
    return targets, coord
```

```python
a = part1(gene_set=hm_human,tf_name='INTERFERON_GAMMA_RESPONSE',
                       group=lupus_cleaned_sub.obs['Celltype'],result=lupus_hm,
                       count=lupus_cleaned_sub)
```

```python
lupus_TF_dca = a[1].rename(columns={0: "dca1", 1: "dca2"})
targets = a[0]
lupus_fit = pd.DataFrame(lupus_cleaned_sub[:,targets].X, columns=targets, index = lupus_cleaned_sub.obs.index.tolist())
lupus_group = lupus_cleaned_sub.obs['Celltype']
```

```python
%R -i lupus_TF_dca,lupus_fit,lupus_group
```

```python
%%R
library(pheatmap)
# Scaled expression of each gene
fit <- t(apply(lupus_fit,2,function(x) (x-mean(x))/sd(x)))
summary(fit[1,])
fit[which(fit > 2)] <- 2
# TF coordinates
tf <- t(lupus_TF_dca)
# Group
group <- lupus_group
# column order
o <- order(group,tf[1,])
# Generate heatmap
anno_col <- data.frame(TF_dca2 = tf[2,o],
                       TF_dca1 = tf[1,o],
                       Group = group[o])
fit_o <- fit[,o]
rownames(anno_col) <- colnames(fit_o)
genes <- names(which(apply(fit_o, 1, function(x) sum(x > 0)) > ncol(fit_o)*0.01))
pheatmap(fit_o[genes,],cluster_rows=T,cluster_cols=F,annotation_col = anno_col,
         show_colnames = F,main="INTERFERON_GAMMA_RESPONSE",
         fontsize_row = 8,width=11,height=8.5,units="inch",
         breaks = seq(from = -2, to = 2, length = 101),
         filename = "INTERFERON_GAMMA_RESPONSE.pdf")
```
